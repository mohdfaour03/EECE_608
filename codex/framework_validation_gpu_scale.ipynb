# GPU-Scale Framework Validation

This notebook launches the sidecar GPU-scale framework validation run from `codex/run_framework_validation_gpu_scale.py`.

Expected outputs:
- `codex/results/framework_validation_gpu_scale/framework_validation_gpu_scale_summary.json`
- `codex/results/framework_validation_gpu_scale/framework_validation_gpu_scale_rows.csv`
- `codex/results/framework_validation_gpu_scale/framework_validation_gpu_scale_checks.json`
- `codex/results/framework_validation_gpu_scale/framework_validation_gpu_scale_report.md`


```python
from pathlib import Path
import os
import sys

def find_repo_root() -> Path:
    candidates = [Path.cwd(), Path.cwd().parent, Path('/kaggle/working')]
    for candidate in candidates:
        if (candidate / 'codex' / 'run_framework_validation_gpu_scale.py').exists():
            return candidate
    raise FileNotFoundError('Could not locate repo root containing codex/run_framework_validation_gpu_scale.py')

REPO_ROOT = find_repo_root()
SCRIPT_PATH = REPO_ROOT / 'codex' / 'run_framework_validation_gpu_scale.py'
RESULTS_DIR = REPO_ROOT / 'codex' / 'results' / 'framework_validation_gpu_scale'

print('python:', sys.executable)
print('repo_root:', REPO_ROOT)
print('script_path:', SCRIPT_PATH)
print('results_dir:', RESULTS_DIR)
os.chdir(REPO_ROOT)
```

```python
import torch

print('cuda_available =', torch.cuda.is_available())
print('device_count =', torch.cuda.device_count())
if torch.cuda.is_available():
    print('device_name =', torch.cuda.get_device_name(0))
else:
    raise RuntimeError('CUDA is not available in this notebook session. Enable a GPU accelerator before running the validation.')
```

```python
import subprocess
import sys

cmd = [sys.executable, str(SCRIPT_PATH)]
print('Running:', ' '.join(cmd))
completed = subprocess.run(cmd, cwd=str(REPO_ROOT), check=True)
print('returncode =', completed.returncode)
```

```python
from pathlib import Path

expected = [
    RESULTS_DIR / 'framework_validation_gpu_scale_summary.json',
    RESULTS_DIR / 'framework_validation_gpu_scale_rows.csv',
    RESULTS_DIR / 'framework_validation_gpu_scale_checks.json',
    RESULTS_DIR / 'framework_validation_gpu_scale_report.md',
]

for path in expected:
    print(path, 'exists=' , path.exists())
```
